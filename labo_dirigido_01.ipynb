```r
# LEY DE LOS GRANDES NUMEROS

N <- 200 # Tamaño de la muestra
# poblacion uniforme (0,1)

x1 <- runif(N) # UNA v. a. con distribucion uniforme en (0,1), su media sera 0.5
# rnorm(N)

x1_raya <- cumsum(x1)/(1:N)
# cumsum: suma acumulativa (primera obs, primera + segunda, etc)
# entonces, x1_raya es la sucesion de realizaciones de las medias muestrales
# (aumentando la cantidad de observaciones)

plot(1:N, x1_raya, ylim=c(0,1), type='l')
```

## 7. Ejecutar codigo y explicar

```r
# rep("f", 91) #Crea un vector de 91 elementos con elementos de tipo 'character' y con valores repetidfos 'f'.
# c(rep("f",91), rep("m",92)) # Luego, se concatena con un vector similar de 92 elementos y con valores repetidos 'm'
# Luego se le aplica la función factor para codificar el vector como una categoría, donde los posibles valores son 'f' o 'm'.
genero <- factor(c(rep("f", 91), rep("m", 92)))
table(genero) #Finalmente, se presenta en una tabla de dos columunas (categorias): 'f' y 'm' cada uno con el respectivo valor de su recurrencia en el vector.
```

```r
# Se reutiliza la variable 'genero' y se codifica de manera explícita con los niveles con los que deseamos clasificar los valores del vector 'genero'
# No existe alguna modificación en comparación a la ejecución anterior pues si no se especifica, la funcion 'factor' define los niveles de acuerdo a la data contenida en la variable.
genero <- factor(genero, levels=c("m", "f"))
table(genero) # se vuelve a presentar una tabla
```

```r
# Se reutiliza la variable 'genero', para definir un nuevo factor pero esta vez tiene niveles de categoría pre-definidos (i.e. 'M', 'f')
genero <- factor(genero, levels=c("M", "f")) # En este caso, solo un valor pre-definido coincide con algunos elementos del vector 'genero', los que no coinciden son reemplazados por <NA>
table(genero) # Se muestra en una tabla la recurrencia de cada nivel de factor, vemos que para el caso del nivel 'M' no hay alguna recurrencia.
```

```r
rm(genero) #Este comando elimina el valor en memoria que estaba almacenado en la variable "genero"
```
